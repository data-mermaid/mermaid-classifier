For this example, ensure that:

1. You have an [MLflow tracking server](https://www.mlflow.org/docs/latest/ml/tracking/server/) up and running

1. You have feature-extractor weights compatible with PySpacer

1. The `.env` file is modified to at least have the following configured according to your setup:

    - `MLFLOW_TRACKING_SERVER`
    - `WEIGHTS_LOCATION`
    - `SPACER_EXTRACTORS_CACHE_DIR`

Then, this cell should do a training run:

```python
from mermaid_classifier.pyspacer.train import run_training

run_training(
    # Training will proceed once 10000 training annotations
    # are collected.
    # This way we limit the training time/resources
    # for testing purposes.
    annotation_limit=10000,
)
```

The print output of that training run should include a line like:

> Model ID: m-f9e4bd067e4f4517962f3768c7aee106

Copy the model ID from this line, and use it in the cell below. This cell should show a MERMAID benthic image with points classified using this model:

```python
from mermaid_classifier.pyspacer.annotation import AnnotationRun

annotation_run = AnnotationRun(
    image='s3://coral-reef-training/mermaid/0032dba6-8357-42e2-bace-988f99032286.png',
    points_csv='0032dba6_points.csv',
    # Replace this with the model ID you got from training.
    classifier='m-f9e4bd067e4f4517962f3768c7aee106',
)
annotation_run.show()
```
